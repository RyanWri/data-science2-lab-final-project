```python
# Convert time cells to datatime pandas format for future processing

df_with_time = erBeforeHospitalization2_df.copy()

df_with_time['Admission_Entry_Date'] = pd.to_datetime(df_with_time['Admission_Entry_Date'])
df_with_time['Release_Date'] = pd.to_datetime(df_with_time['Release_Date'])

df_with_time['ev_Admission_Date'] = pd.to_datetime(df_with_time['ev_Admission_Date'])
df_with_time['ev_Release_Time'] = pd.to_datetime(df_with_time['ev_Release_Time'])

df_with_time['Admission_Entry_Date2'] = pd.to_datetime(df_with_time['Admission_Entry_Date2'])
df_with_time['Release_Date2'] = pd.to_datetime(df_with_time['Release_Date2'])


	
```

#### Fill null values and data cleaning

```python
# By looking at ("ev_Admission_Date", "ev_Release_Time") we can see that the dates are almost identical to ("Admission_Medical_Record2", "Release_Date2")
# Thus, I will replace these dates where the dates are missing:

# Replace null values in 'ev_Admission_Date' with values from 'Admission_Medical_Record2'
df_with_time['ev_Admission_Date'].fillna(df_with_time['Admission_Medical_Record2'], inplace=True)

# Replace null values in 'ev_Release_Time' with values from 'Release_Date2'
df_with_time['ev_Release_Time'].fillna(df_with_time['Release_Date2'], inplace=True)


# The "Admission_Medical_Record" and "Medical_Record" columns does not give much information so I will drop them.
df_with_time = df_with_time.drop(columns=['Admission_Medical_Record', 'Medical_Record'])

```

```python

unique_getting_to_er= df_with_time['דרך הגעה למיון'].unique()
unique_er = df_with_time['מיון'].unique()
unique_urgencyLevelTime = df_with_time['urgencyLevelTime'].unique()
unique_codeDoctor = df_with_time['codeDoctor'].unique()

# To print the unique values

print(unique_getting_to_er, "\n\n")
print(unique_er, "\n\n")
print(unique_urgencyLevelTime, "\n\n")
print(unique_codeDoctor, "\n\n")
```

```python
# Drop rows that have many null values in the same row

columns_to_check = ["codeDoctor", "אבחנות במיון", "urgencyLevelTime", "מיון", "דרך הגעה למיון"]
df_with_time = utils.drop_rows_with_many_nans(df_with_time, columns_to_check, max_nans=2)
utils.check_null_values_in_df(df_with_time)
```

```python
df_encoded = df_with_time.copy()
df_encoded = pd.get_dummies(df_encoded, columns=['דרך הגעה למיון'])
df_encoded = pd.get_dummies(df_encoded, columns=['מיון'])
```

```python
getting_to_er_and_emergency = [col for col in df_encoded.columns if col.startswith('דרך הגעה למיון')]
er_department_and_emergency = [col for col in df_encoded.columns if col.startswith('מיון')]



pair_counts1 = pltu.count_pairs(df_encoded, 'urgencyLevelTime', getting_to_er_and_emergency)

# Apply a color gradient to the pair_counts DataFrame
styled_table = pair_counts1.style.background_gradient(cmap='coolwarm').set_caption("Urgency vs Way of Arrival")
styled_table
```

```python

pair_counts2 = pltu.count_pairs(df_encoded, 'urgencyLevelTime', er_department_and_emergency)
# Apply a color gradient to the pair_counts DataFrame
styled_table = pair_counts2.style.background_gradient(cmap='coolwarm').set_caption("Urgency vs ER department")
styled_table
```

```python

pair_counts3 = pltu.count_active_pairs(df_encoded, er_department_and_emergency, getting_to_er_and_emergency)

# Apply a color gradient to the pair_counts DataFrame
styled_table = pair_counts3.style.background_gradient(cmap='coolwarm').set_caption("ER department vs Way of Arrival")
styled_table

```

We can see from the previous tables that most of the ways of arrival  to the hospital and most of the ER departments relate mainly to urgency level 4, and some of them to level 3. 
In addition, we can see that most of the ways of arrival relates to "מיון פנימי", Arriving alone relates also to "רפואה דחופה וזיהומים" and arriving with family member or with ambulance relates as well to "מיון מהלכים" .

Thus, I will replace every column where there is no urgency level with urgency level 4. 
In addition, every column without ER department will be replaced with "מיון פנימי". 
We can see from the first table that most of the patients arrived alone or in ambulance or with a family member, thus, I will put a random choice for the computer to decide the way of arrival.

```python
df_with_time['urgencyLevelTime'].fillna(4, inplace=True)

df_with_time['דרך הגעה למיון'].fillna(utils.choose_random_string("דרך הגעה למיון_נהג אמבולנס", "דרך הגעה למיון_לבד", "דרך הגעה למיון_בן משפחה"), inplace=True)

# df_with_time['מיון'].fillna(("מיון פנימי"), inplace=True)
```
